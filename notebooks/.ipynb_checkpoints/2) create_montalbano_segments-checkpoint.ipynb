```python
'''
- Isolate the Montalbano sequences, i.e., identify the start and end of each performed gesture. The isolation annotations are available in 'montalbano_segments.csv' 
- expected folder structure: the montalbano source frames should be arranged as follows:
frame_source:
 - training:
   - Samplexxxxx_color
       - 0.jpg
       - 1.jpg
       - 2.jpg
       - ...
   - Samplexxxxx_color
 - test:
   - Samplexxxxx_color
   - Samplexxxxx_color
 - validation:
   - Samplexxxxx_color
   - Samplexxxxx_color

Notes:
    - this step is required only for Montalbano. The GRIT sequences are already isolated
'''
```

```python
df = pd.read_csv('montalbano_segments.csv', dtype=object, sep=",", header=None)
print(df[df[0]=='00659'])
```

```python
path = '/path/to/montalbano_frame_source' #path to dataset (frames)
for folder in os.listdir(path): #training, test, etc.
    print(folder)
    for subfolder in os.listdir(os.path.join(path,folder)): #Sample00021_color, Sample00022_color, etc.
        if (subfolder.startswith(str(""))): #use for testing
            #print(subfolder)
            sample_index = subfolder[6:-6]
            segments = df[df[0]==sample_index]
            #print(len(segments))
            if (len(segments) > 0):
                target_path = os.path.join(path, folder, subfolder)
                print("target_path ", target_path)
                for index, row in segments.iterrows():
                    start_file, end_file = row[1].split()[0], row[1].split()[1]
                    os.rename(os.path.join(target_path, start_file+".jpg"), os.path.join(target_path, start_file+"-start.jpg"))
                    os.rename(os.path.join(target_path, end_file+".jpg"), os.path.join(target_path, end_file+"-end.jpg"))
                    print(row[0], row[1])
                    #print(start + ' ' + end)
```
