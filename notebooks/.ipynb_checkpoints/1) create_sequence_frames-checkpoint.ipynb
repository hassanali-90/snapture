```python
import os
import cv2
```

```python
'''
- this step will convert Montalbano video source into frames
- expected folder structure: the montalbano source videos should be arranged as follows:
video_source:
 - training:
   - Samplexxxxx_color.mp4
   - Samplexxxxx_color.mp4
 - test:
   - Samplexxxxx_color.mp4
   - Samplexxxxx_color.mp4
 - validation:
   - Samplexxxxx_color.mp4
   - Samplexxxxx_color.mp4
   
- Notes:
    - this step is required for Montalbano only since GRIT is already available in frame-by-frame format
    - videos ending with '_color' represents RGB data (used in this experiment). Other video files can (and should) be safely removed
    - make sure you have plenty of free disk space!
'''
```

```python
path = '/path/to/video_source'
target_path = '/path/to/montalbano_frame_source' #make sure this folder exits
for folder in os.listdir(path): #test, train etc.
    target_folder = os.path.join(target_path, folder)
    if not os.path.exists(target_folder):
        os.mkdir(target_folder)
    for file in os.listdir(os.path.join(path,folder)): #Sample00021_color, Sample00022_color, etc.
        if (file.startswith(str(""))): #well that's all of 'em
            video = cv2.VideoCapture(os.path.join(path, folder, file))
            success, frame = video.read()
            count = 0
            while success:
                print(os.path.join(target_path, folder,os.path.splitext(file)[0], str(count) + '.jpg'))
                target_subfolder = os.path.join(target_path, folder,os.path.splitext(file)[0])
                if not os.path.exists(target_subfolder):
                    os.mkdir(target_subfolder)
                print(cv2.imwrite(os.path.join(target_subfolder, str(count) + '.jpg'), frame))
                success, frame = video.read()
                print('Read a new frame: ', success)
                count += 1
```
